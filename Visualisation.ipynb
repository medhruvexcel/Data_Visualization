### Basic charts 

#### Line chart 

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]


pt.plot(x, y, marker='o', linestyle='-', color='r')
pt.title('Line Chart')
pt.xlabel('X-axis')
pt.ylabel('Y-axis')

pt.show()
```

#### Bar chart 

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]


pt.bar(categories, values, color='skyblue', width =0.5)
pt.title('Bar Chart')
pt.xlabel('Categories')
pt.ylabel('Values')
pt.show()
```

#### Pie Chart

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]


pt.pie(sizes, labels=labels, autopct='%0.1f%%')
pt.title('Pie Chart')
pt.show()
```

### Histo Question

```python

scores = [22, 45, 60, 75, 80, 95, 55, 65, 70, 85, 90, 72, 88, 76, 80, 58, 62, 73, 68, 78]


pt.hist(scores, bins=10, color ="grey", edgecolor='black', alpha=0.7)
pt.title('Distribution of Exam Scores')
pt.xlabel('Score Range')
pt.ylabel('Number of Students')

pt.show()
```

### Subplot

```python
import matplotlib.pyplot as plt 
```

```python
stud= ['john ' ,'dhruv','adarsh ',' rose']
math_scores=[85,67,78,90]
science_scores=[78,93,78,84]

#creating subplot 
fig , axs =plt.subplots(1,2,figsize=(8,4))
#plotting thw scores 
axs[0].bar(stud,math_scores, color='b' )
axs[0].set_title('Math Scores')
axs[0].set_xlabel('Students')
axs[0].set_ylabel('Score')

#plotting science score 
axs[1].bar(stud,science_scores , color='r' )
axs[1].set_title('Science Scores')
axs[1].set_xlabel('Students')
axs[1].set_ylabel('score')
plt.tight_layout()
plt.show()
```

```python
x=[]
for i in range (10 , 110 , 10 ):
    x.append(i)
y=[]
for i in range(1 , 11 ):
    y.append(i)

plt.subplot(2,2,1)
plt.plot(x, y, color='r',marker ='*')

plt.subplot(2,2,2)
plt.pie(y,labels=x )

plt.subplot(2,2,4)
plt.bar(x, y)

plt.subplot(2,2,3)
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%0.1f%%')
```

```python
x=[]
for i in range (10 , 110 , 10 ):
    x.append(i)
y=[]
for i in range(1 , 11 ):
    y.append(i)

plt.subplot(1,4,1)
plt.plot(x, y, color='r',marker ='*')

plt.subplot(1,4,2)
plt.pie(y,labels=x )

plt.subplot(1,4,3)
plt.bar(x, y)

plt.subplot(1,4,4)
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%0.1f%%')
```

### Formatting plot

### text plot

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]

plt.plot(x,y)
plt.title('plot with general text')
plt.xlabel('X axis ')
plt.ylabel('Y axis ')

plt.text(2,8,'Language', fontsize=10, color='white' , bbox={'facecolor' : 'lightblue'})
plt.annotate(r'java', xy=(4,7), xytext=(2,10), arrowprops=dict(facecolor='black',shrink=0.5), fontsize=12, color='red')
plt.show()
```

### Xlim , Ylim 

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.plot(x,y)
plt.xlim(0,7)
plt.ylim(0,12)
```

```python
x = [1, 2, 3, 4, 5] 
y = [2, 3, 5, 7, 11]



# Creating plot plt.plot(x, y)

#Customizing ticks
plt.plot(x,y)

plt.xticks([1, 2, 3, 4, 5], ['One', 'Two', 'Three', 'Four', 'Five'])

plt.yticks([2, 4, 6, 8, 10, 12], ['Two', 'Four', 'Six', 'Eight', 'Ten', 'Twelve'])

plt.title('Custom Ticks Example')

plt.xlabel('X-axis')

plt.ylabel('Y-axis')

#Showing plot

plt.show()
```
