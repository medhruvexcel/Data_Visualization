### Task 1: Visualize the daily temperature changes over time in a city 

```python
import matplotlib.pyplot as plt
```

```python
days = list(range(1, 32))
temperature = [22, 23, 24, 26, 25, 27, 28, 29, 30, 29, 28, 26, 25, 24, 23, 22, 24, 26, 27, 29, 30, 32, 33, 34, 35, 33, 32, 31, 29, 27, 25]
```

```python
# Plotting the temperature over days
plt.figure(figsize=(15, 5))
plt.plot(days, temperature, marker='o', color='blue', linestyle='-', linewidth=1)
plt.title('Daily Temperature Changes Over Time in a City')
plt.xlabel('Days')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
#conclusion 
```

```python
"""The temperature generally increases during the first half of the month, reaching slight early a peak mid-month.
After that, there’s a gradual decline, with minor fluctuations towards the mid of the month.\nThen there rise temperature of peak 35 degrees celcius on 25th day 
This pattern may suggest seasonal changes or variability due to weather conditions."""
```

### Task 2: Create a Line Plot to Visualize Daily Closing Prices of a Stock Over a Year

```python
days = list(range(1, 78))
stock_prices = [250, 255, 260, 265, 270, 268, 272, 275, 280, 285, 290, 295, 300, 305, 310, 308, 312, 315, 320, 325, 330, 328, 335, 340, 345, 350, 355, 360, 365, 370, 368, 372, 375, 380, 385, 390, 395, 392, 388, 385, 380, 375, 370, 365, 360, 355, 350, 345, 340, 335, 330, 325, 320, 315, 310, 305, 300, 295, 290, 285, 280, 275, 270, 265, 260, 255, 250, 245, 240, 235, 230, 225, 220, 215, 210, 205, 200]
```

```python
#Plotting the stock prices over days
plt.figure(figsize=(25, 6))
plt.plot(days, stock_prices, marker='o', color='red', linestyle='-', linewidth=2)
plt.title('Daily Closing Prices of a Stock Over a Year')
plt.xlabel('Days')
plt.ylabel('Stock Prices ($)')
plt.grid(True)
plt.show()
```

```python
"""The stock price initially shows a consistent upward trend, reaching a peak around the 35th day. Following this, there’s a gradual decline, indicating a possible market correction or shift in investor sentiment. The fluctuations towards the end might reflect market volatility or external factors affecting the stock."""
```
