### Task 1 : Analyze the relationship between the size of houses (measured in square footage) and their selling prices in a particular neighborhood. You have collected data on various houses in that neighborhood

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Given data
square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800, 3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])

# Creating a scatter plot
plt.figure(figsize=(15, 6))
plt.scatter(square_footage, selling_prices, color='red', marker='o')

# Adding title and labels
plt.title('Relationship Between House Size and Selling Prices')
plt.xlabel('Square Footage')
plt.ylabel('Selling Prices (in $1000)')
plt.grid(True)
plt.show()
```

```python
"""The scatter plot reveals a positive correlation between the size of houses (measured in square footage) and their selling prices. As the square footage increases, the selling price tends to increase as well. The relationship appears to be linear, suggesting that larger houses in this neighborhood are sold at higher prices. This trend indicates that house size is a significant factor in determining the selling price, and buyers are willing to pay more for larger homes."""
```

### Task 2 :  Create a Pie Chart to Visualize Monthly Income Distribution

```python
# Given data
income_sources = ['Salary', 'Freelance', 'Investments', 'Rental', 'Other']
monthly_income = [5000, 1500, 1000, 600, 400]

# Creating the pie chart
plt.figure(figsize=(8, 8))
plt.pie(monthly_income, labels=income_sources, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])

# Adding title
plt.title('Distribution of Monthly Income by Source')
plt.show()
```

```python
"""The pie chart shows that the majority of your monthly income comes from your salary, which constitutes the largest portion at 58.8%. Freelance work contributes 17.6%, making it the second-largest source of income. Investments and rental income provide 11.8% and 7.1% respectively, while other income sources contribute a small portion of 4.7%. This distribution suggests that your financial stability heavily relies on your salary, with freelance work as a notable secondary source. The smaller contributions from investments and rental income indicate some diversification, though the majority of your income is still tied to your primary job and freelance activities."""
```

### Task 3 :  illustrate the distribution of a company's revenue across its various business segments using pie chart.

```python
# Data for the pie chart
segments = ['Product A', 'Product B', 'Services', 'Licensing']
revenue_percentages = [45, 25, 15, 15]

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(revenue_percentages, labels=segments, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])

# Title of the pie chart
plt.title('Revenue Distribution Across Business Segments')

# Show the pie chart
plt.show()
```

```python
# '''The pie chart visualizes the revenue distribution across the company's business segments. From the chart, we can see that:'''
# '''Product A contributes the largest share of revenue, accounting for 45%.'''
# '''Product B follows with 25% of the revenue.'''
# '''Services and Licensing each contribute 15% to the total revenue.'''
# '''This distribution highlights the company's strong reliance on Product A for revenue, while other segments contribute a smaller portion. Understanding this distribution can help in strategic decision-making, such as where to focus marketing efforts or consider expanding product lines.'''
```

### Task 4 : Suppose you're a sales manager for an e-commerce company, and you want to create a figure with subplots to compare the sales performance of different product categories over time. You have sales data for four product categories: Electronics, Clothing, Home & Garden, and Sports & Outdoors.  

```python
# Data for sales performance
months = np.arange(1, 13)
electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000, 36000, 38000, 39000, 41000, 42000])
clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000])
home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000])

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Electronics Sales
axs[0, 0].plot(months, electronics_sales, marker='o', color='blue')
axs[0, 0].set_title('Electronics Sales')
axs[0, 0].set_xlabel('Month')
axs[0, 0].set_ylabel('Sales ($)')
axs[0, 0].grid(True)

# Clothing Sales
axs[0, 1].plot(months, clothing_sales, marker='o', color='green')
axs[0, 1].set_title('Clothing Sales')
axs[0, 1].set_xlabel('Month')
axs[0, 1].set_ylabel('Sales ($)')
axs[0, 1].grid(True)

# Home & Garden Sales
axs[1, 0].plot(months, home_garden_sales, marker='o', color='red')
axs[1, 0].set_title('Home & Garden Sales')
axs[1, 0].set_xlabel('Month')
axs[1, 0].set_ylabel('Sales ($)')
axs[1, 0].grid(True)

# Sports & Outdoors Sales
axs[1, 1].plot(months, sports_outdoors_sales, marker='o', color='purple')
axs[1, 1].set_title('Sports & Outdoors Sales')
axs[1, 1].set_xlabel('Month')
axs[1, 1].set_ylabel('Sales ($)')
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
'''Conclusion
The subplots illustrate the sales performance of different product categories over the course of a year:'''
```

```python
'''Electronics: Sales show a steady increase throughout the year with a noticeable peak in the last two months, indicating strong performance, especially in the holiday season.'''
```

```python
'''Clothing: Sales exhibit a consistent upward trend, suggesting steady growth and possibly successful marketing strategies or seasonal promotions.'''
```

```python
'''Home & Garden: This category shows a steady increase in sales, similar to clothing, indicating a robust market presence and consistent consumer interest'''
```

```python
'''Sports & Outdoors: Sales also demonstrate a gradual increase over time, but the growth is slower compared to the other categories, possibly indicating a need for enhanced marketing or seasonal promotions.'''
```

```python
'''Overall, all categories show positive growth, but Electronics and Home & Garden categories stand out with stronger performance trends.'''
```
